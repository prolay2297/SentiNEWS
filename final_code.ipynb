```python
# Latest
import re
import nltk
from indicnlp.tokenize import indic_tokenize
from pywebio.input import input, select
from pywebio.output import put_text, put_markdown, put_buttons, put_image, put_link, put_row, clear, put_table
import serpapi
import newspaper
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from gtts import gTTS
from io import BytesIO
import pygame
from newspaper import Article
from textblob import TextBlob
import pandas as pd
from diffusers import StableDiffusionPipeline
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import torch
from torch.nn import Module, Linear
from torch.optim import SGD
from rouge import Rouge
from transformers import pipeline
import pandas as pd
from pywebio.output import put_image

# Download NLTK stopwords data
nltk.download('stopwords')

def display_image(image_path):
    with open(image_path, 'rb') as f:
        image_data = f.read()
        put_image(image_data, width='50%')

# SerpApi key
serpapi_key = 'f83bb6deae3e871b762d58bdf4143adbc89f8d17dadd51de218df30478ab38ae'
client = serpapi.Client(api_key=serpapi_key)
hindi_stopwords = set(nltk.corpus.indian.words('hindi.pos'))
df = pd.DataFrame(columns=['Text', 'Language', 'ROUGE Scores'])

# Initialize the summarization pipeline with max_length parameter
summarizer = pipeline("summarization")

# Instantiate the ROUGE scorer
rouge = Rouge()


def process_text(text, language):
    text = text.lower()
    text = re.sub('((www.\S+)|(https?://\S+))', '', text)
    text = re.sub('@\S+', '', text)
    text = re.sub('[\s]+', ' ', text)
    text = re.sub(r'#(\S+)', r'\1', text)
    text = re.sub(r'[.!:?-]', '', text)
    text = text.strip('"\'')
    return text

def tokenization(text, language):
    tokens = []
    if language == 'Hindi':
        for t in indic_tokenize.trivial_tokenize(text):
            tokens.append(t.strip('\n'))  # Remove '\n' from each token
    else:
        tokens.extend(nltk.word_tokenize(text))
    return tokens

def remove_stopwords(text, language):
    if language == 'hi':
        # Use Hindi stopwords
        stop_words = hindi_stopwords
    else:
        # Use English stopwords
        stop_words = set(stopwords.words('english'))
    word_tokens = nltk.word_tokenize(text)
    filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
    return ' '.join(filtered_text)

def stop_audio():
    # Stop the audio playback
    pygame.mixer.music.stop()  

def generate_image(prompt):
    model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"
    pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16)
    pipe = pipe.to("cuda")

    image = pipe(prompt).images[0]
    return image

def play_audio(text, language):
    if not text:
        put_text("No text to speak.")
        return

    # Generate audio using gTTS
    tts = gTTS(text=text, lang=language)
    fp = BytesIO()
    tts.write_to_fp(fp)
    fp.seek(0)

    # Play the audio using pygame
    pygame.mixer.init()
    pygame.mixer.music.load(fp)
    pygame.mixer.music.play()
    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)

def analyze_sentiment(values):
    # Determine sentiment label
    if values >= 0.05:
        return 'Positive'
    elif values <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
    
def play_audio_closure(summary, lang):
    def play_audio_inner(btn_val):
        play_audio(summary, lang)
    return play_audio_inner
    
def play_button(idx,df):
    if idx==1:
        put_buttons(['Play Audio', 'Stop Audio'], onclick=[lambda btn_val=None: play_audio(df['Text'][idx], df['Language'][idx]), lambda btn_val=None: stop_audio()])
    elif idx==2:
        put_buttons(['Play Audio', 'Stop Audio'], onclick=[lambda btn_val=None: play_audio(df['Text'][idx], df['Language'][idx]), lambda btn_val=None: stop_audio()])
    elif idx==3:
        put_buttons(['Play Audio', 'Stop Audio'], onclick=[lambda btn_val=None: play_audio(df['Text'][idx], df['Language'][idx]), lambda btn_val=None: stop_audio()])
    elif idx==4:
        put_buttons(['Play Audio', 'Stop Audio'], onclick=[lambda btn_val=None: play_audio(df['Text'][idx], df['Language'][idx]), lambda btn_val=None: stop_audio()])
    elif idx==5:
        put_buttons(['Play Audio', 'Stop Audio'], onclick=[lambda btn_val=None: play_audio(df['Text'][idx], df['Language'][idx]), lambda btn_val=None: stop_audio()])

def calculate_rouge(summary, reference):
    # Calculate ROUGE score
    scores = rouge.get_scores(summary, reference)
    return scores

def print_rouge_scores(idx, rouge_scores):
    put_table(rouge_scores)

def get_news():
    user_query = input("Enter your news query:", type="text")
    language = select('Select Language', ['English', 'Hindi'])  # Add other languages if available
    return user_query, language

def get_keywords(query, language):
    keywords = []
    processed_query = process_text(query, language)
    tokens = tokenization(processed_query, language)
    # Add tokens to keywords
    keywords.extend(tokens)
    return keywords

# In your search_and_summarize_news() function, modify the keyword extraction part as follows:
def calculate_rouge_scores():
    rouge_scores_list = []
    for idx, row in df.iterrows():
        if row['ROUGE Scores'] != "":
            rouge_scores_list.append((row['ROUGE Scores'], row['Link']))
    return rouge_scores_list

def print_all_rouge_scores():
    # Print ROUGE scores for all articles
    for idx in range(1, 6):  # Loop through the indices of the DataFrame
        put_table(df['ROUGE Scores'][idx])
        put_link("Article Link", df['Link'][idx])
def print_rouge_scores(idx):
    put_table(df['ROUGE Scores'][idx])
    #put_link("Article Link", df['Link'][idx])

# Create a DataFrame to store summaries and ROUGE scores
df = pd.DataFrame(columns=['Text', 'Language', 'ROUGE Scores', 'Link'])

def search_and_summarize_news():
    while True:

        image_path = 'Logo.png'  

        # For displaying the image
        display_image(image_path)
        user_query, language = get_news()
        language_code = 'en' if language == 'English' else 'hi'  # Add other languages if available
        # Get keywords
        keywords = get_keywords(user_query, language)

        if language_code=='hi':
            user_query=user_query+" "+"in hindi news"
        
        

        # Perform the search using SerpApi
        result = client.search(
            q=user_query,
            engine="google",
            location="Delhi, India",
            hl="hi" if language == 'Hindi' else "en",
            gl="in"
        )

        put_markdown("### Summarized News:")
        for idx, item in enumerate(result.get("organic_results", [])[:5], 1):  # Display only top 5 relevant links
            try:
                article = Article(item['link'], language=language_code)  # Specify language
                article.download()
                article.parse()
                article.nlp()
                
                # Add keyword tokens from article to keywords
                keywords.extend(article.keywords[:5])

                # Remove stopwords from summary
                filtered_summary = remove_stopwords(article.summary, language_code)  # Use language-specific stopwords

                put_markdown(f"#### {item['link']}")
                put_text(f"Title: {article.title}")

                # Generate summary, analyze sentiment, and add button for audio playback
                if language == 'English':
                    summarized_text = summarizer(article.summary)[0]['summary_text']
                else:
                    summarized_text = article.summary

                # Calculate and store ROUGE scores
                rouge_scores = calculate_rouge(summarized_text, article.text)
                df.loc[idx] = [summarized_text, language_code, rouge_scores, item['link']]  # Store summary and ROUGE scores

                put_text(f"Summary: {summarized_text}")
                
                if language == 'English':
                    put_text(f"Top 5 Keywords: {keywords[:5]}")
                else:
                    indic_string = process_text(article.title, language)
                    processed_tokens = tokenization(indic_string, language)
                    put_text(f"Top 5 Keywords: {article.keywords[:5]}")
                        
                # Calculate sentiment using TextBlob
                blob = TextBlob(article.text)
                sentiment = blob.sentiment.polarity
                sentiment_label = analyze_sentiment(sentiment)
                put_text(f"Sentiment: {sentiment_label}")
                
                play_button(idx, df)
                
                # Print ROUGE scores for each article along with their links
                put_buttons(["Print ROUGE Scores"], onclick=lambda btn_val=None, idx=idx: print_rouge_scores(idx))
                
            except newspaper.ArticleException as e:
                put_markdown(f"Error processing article: {e}")
                continue

            prompt = " ".join(keywords)
            if language == 'English':
                prompt = " ".join(keywords)
            else:
                prompt = " ".join(processed_tokens)
            image = generate_image(prompt)
            put_image(image, width='50%')

            put_row([
                put_link("Read more", item['link'])
            ])

            put_text('-' * 80)

        choice = select("Do you want to search again?", ['Yes', 'No', 'Clear'])
        if choice == 'No':
            break
        elif choice == 'Clear':
            clear()  # Clear the output before next search

#search_and_summarize_news()
search_and_summarize_news()


#put_buttons(["Print All ROUGE Scores"], onclick=print_all_rouge_scores)

#search_and_summarize_news()
```
